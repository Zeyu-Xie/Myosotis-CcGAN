This notebook shows how to load and process images.

First import some packages.

```python
import h5py
import numpy as np
from PIL import Image
```

## Load an image from a dataset

Load the dataset and get the `numpy` array of a certain image.

```python
# Path
dataset_path = "/home/ubuntu/Desktop/Ra/datasets/Ra_128_indexed.h5"

# Load `numpy` array
with h5py.File(dataset_path, "r") as f:
    images = f["images"][:]
    labels = f["labels"][:]

# Get the 100-th `numpy` array
image = images[100]
label = labels[100]
print(f"Image shape: {image.shape}")
print(f"Label: {label}")
```

## Load the `Image` object and use it

After we have the `numpy` array, we can load it as an `Image` object from `pillow` and use it.

```python
# Load the `numpy` array as an `Image` object
image_obj = Image.fromarray(image)

# The image looks like this
image_obj
```

## Write an image into a `numpy` array

Just use `np.array()` function, and we can write an image into a `numpy` array.

```python
image = np.array(image_obj)
print(image.shape)
```
